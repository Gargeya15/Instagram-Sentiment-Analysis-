Data cleanig using NLP

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z ]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

Applying changes 

```python
df['clean_comment'] = df['comment'].apply(clean_text)
```

Let us start with sentiment analysis 

Here i will take each comment into any 3 of below 
Positive 
Negative 
Netural

```python
sia = SentimentIntensityAnalyzer() #Initialization of model 
```

```python
def get_sentiment(text):
    score = sia.polarity_scores(text)['compound']  
    
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"
    
```

```python
df['sentiment']=df['clean_comment'].apply(get_sentiment)
```
